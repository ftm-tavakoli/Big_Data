```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# C.

```python
y1 = pd.read_csv('/content/Y1.csv').to_numpy().T
y2 = pd.read_csv('/content/Y2.csv').to_numpy().T
```

```python
c1=np.cov(y1)
c2=np.cov(y2)
U1, S1, Vh1 = np.linalg.svd(c1, full_matrices=True)
U2, S2, Vh2 = np.linalg.svd(c2, full_matrices=True)
```

```python
Um = U1
Vhm = Vh2
Sm= pow(S1,0.5)
M = Um.dot(np.diag(Sm)).dot(Vhm)
print('M equal to :\n',M)
```

```python
Umt = U2
Vhmt = Vh1
Smt= pow(S2,0.5)
MT = Umt.dot(np.diag(Smt)).dot(Vhmt)
print('M^T equal to :\n',MT)
```

# D.

```python
mean = [0, 0]
cov = [[1, 0], [0, 1]]  # diagonal covariance

x= np.random.multivariate_normal(mean, cov,50000).T
```

```python
Y1_hat = M.dot(x)
```

```python
plt.scatter(y1_hat[0], y1_hat[1],s=2, label='Y1_hat',c ='c')
plt.scatter(y1[0], y1[1],s=2, label='Y1', c='r')
plt.legend()
plt.show()
```

```python
Y2_hat = MT.dot(x)
```

```python
plt.scatter(Y2_hat[0],Y2_hat[1],s=2, label='Y2_hat',c ='c')
plt.scatter(y2[0], y2[1],s=2, label='Y2',c='r')
plt.legend()
plt.show()
```
