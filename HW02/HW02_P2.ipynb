```python
class BloomFilter:
    def __init__(self, n, m, k, hash_func):
        self.n = n
        self.m = m  # Calculate the size of the bit array
        self.k = k
        self.hash_func = hash_func
        self.bit_array = [False] * self.m  # Initialize the bit array with 0 values

    def add(self, item):
        for i in range(self.k):
            index = self.hash_func(item, i) % self.m  # Get the index for each hash function
            self.bit_array[index] = True  # Set the corresponding bit to 1
        return True  # Item added successfully

    def test(self, item):
        hash_results = []
        for i in range(self.k):
            index = self.hash_func(item, i) % self.m  # Get the index for each hash function
            hash_results.append(self.bit_array[index])
        if all(hash_results):
            return True
        else:
            print('{} is not in the dataset'.format(item))
            return False

    def get_bit_array(self):
        return self.bit_array


def hash_func_1(item, i):
    c = [ord(ch) for ch in item]
    # Select prime number from the list based on the hash function index
    p = primes[i % len(primes)]  
    l = len(c)
    min_c = min(c)
    max_c = max(c)
    index = (min_c + math.prod(c)) * p + ((sum(c)) * pow(p, (l // 2))) + (max_c * pow(p, (l - 1)))
    return index


def hash_func_2(item, i):
    c = [ord(ch) for ch in item]
    p = primes[i % len(primes)]
    l = 0
    index = 0 
    for x in c :
      index = index + (x *pow(p,l))
      l = l+1
    return index 



```
